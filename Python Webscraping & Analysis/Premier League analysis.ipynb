```python
combined_data = pd.concat([df1, df2, df3], ignore_index=True)
print(combined_data)
```

```python
# Lets explore our data
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Everything looks good except for the date and time columns, lets take a look and change them.
print(combined_data.head())
print(combined_data.tail())
```

```python
combined_data['Converted_Date'] = pd.to_datetime(combined_data['Date'], format = '%m/%d/%Y')
print('Converted Date column')
print(combined_data)


print(combined_data.dtypes)
```

```python
combined_data['Year'] = combined_data['Converted_Date'].dt.year
print('Extracted Year: ')
print(combined_data[['Date', 'Converted_Date', 'Year']])
```

```python
combined_data['Score'] = combined_data['Score'].str.replace('\x96', '-', regex=False)
```

```python
# We can use str.split to extract the actual goals scored for the home and away teams in the game.
combined_data['Home_Score'] = combined_data['Score'].str.split('-').str[0].astype(int)
combined_data['Away_Score'] = combined_data['Score'].str.split('-').str[1].astype(int)
print(combined_data.head())
```

```python
# Home xG and Away xG per team
home_xg = combined_data.groupby('Home')['xG_H'].sum().reset_index()
home_xg.columns = ['Team', 'Total_xG_H']

away_xg = combined_data.groupby('Away')['xG_A'].sum().reset_index()
away_xg.columns = ['Team', 'Total_xG_A']

total_xg = pd.merge(home_xg, away_xg, on='Team', how='outer').fillna(0)

total_xg['Total_xG'] = total_xg['Total_xG_H'] + total_xg['Total_xG_A']

print(total_xg.sort_values(by='Total_xG', ascending=False))
```

```python
# Grouping Total xG by Year 
home_xg = combined_data.groupby(['Year', 'Home'])['xG_H'].sum().reset_index()
home_xg.columns = ['Year', 'Team', 'Total_xG_H']

# Group by Year and Away team to calculate away xG
away_xg = combined_data.groupby(['Year', 'Away'])['xG_A'].sum().reset_index()
away_xg.columns = ['Year', 'Team', 'Total_xG_A']

# Merge the two DataFrames
total_xg = pd.merge(home_xg, away_xg, on=['Year', 'Team'], how='outer').fillna(0)

# Calculate Total xG
total_xg['Total_xG'] = total_xg['Total_xG_H'] + total_xg['Total_xG_A']

# Display the results
print(total_xg.sort_values(by=['Year', 'Total_xG'], ascending=[True, False]))
```

```python
# Which teams have the highest overall average expected goals?
home_perform = combined_data.groupby('Home')['xG_H'].mean().reset_index()
home_perform.rename(columns={'xG_H': 'Average_xG_H'}, inplace=True)
away_perform = combined_data.groupby('Away')['xG_A'].mean().reset_index()
away_perform.rename(columns={'xG_A': 'Average_xG_A'}, inplace=True)

performance = pd.merge(home_perform, away_perform, left_on='Home', right_on='Away', how='outer')
performance.drop(columns='Away', inplace=True)

performance['Overall'] = performance[['Average_xG_H','Average_xG_A']].mean(axis=1)

print(performance)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=performance, x='Average_xG_H', y='Average_xG_A', size='Overall', sizes=(10, 100), legend=None)
for i in range(performance.shape[0]):
    plt.text(performance['Average_xG_H'][i] + 0.05,
             performance['Average_xG_A'][i] + 0.05,
             performance['Home'][i],
             horizontalalignment='right',
             size='medium',
             color='black')
```

```python
# Manchester City and Liverpool dominate the scatterplot, 
# generating upwards of 2.4 xG at home (Liverpool) and 2.0 xG in away games (Manchester City)
```

```python
# Which teams are performing better at home compared to away?
performance['Better_Performance'] = performance.apply(
    lambda row: 'Home' if row['Average_xG_H'] > row['Average_xG_A'] 
                 else ('Away' if row['Average_xG_A'] > row['Average_xG_H'] 
                       else 'Equal'),
    axis=1
)
# Display the results
print(performance[['Home', 'Average_xG_H', 'Average_xG_A', 'Better_Performance']])
```

```python
# Analyzing for over or underperformance on expected goals.

# Grouping home xG
home_xg_cal = combined_data.groupby('Home')['xG_H'].sum().reset_index()
home_xg_cal.rename(columns={'xG_H':'Total_xG_H'},inplace=True)

# Grouping away xG
away_xg_cal = combined_data.groupby('Away')['xG_A'].sum().reset_index()
away_xg_cal.rename(columns={'xG_A':'Total_xG_A'},inplace=True)

# Grouping home actual goals
home_goals = combined_data.groupby('Home')['Home_Score'].sum().reset_index()
home_goals.rename(columns={'Home_Score':'Actual_Goals_H'},inplace=True)

# Grouping away actual goals
away_goals = combined_data.groupby('Away')['Away_Score'].sum().reset_index()
away_goals.rename(columns={'Away_Score':'Actual_Goals_A'},inplace=True)

total_xg_cal= pd.merge(home_xg_cal, away_xg_cal, left_on='Home', right_on='Away', how='outer')
total_goals = pd.merge(home_goals, away_goals, left_on='Home', right_on='Away', how='outer')

performance_summary= pd.merge(total_xg_cal, total_goals, on='Home', how='outer')

performance_summary['Total_xG']= performance_summary['Total_xG_H'].fillna(0) + performance_summary['Total_xG_A'].fillna(0)
performance_summary['Total_Goals'] = performance_summary['Actual_Goals_H'].fillna(0) + performance_summary['Actual_Goals_A'].fillna(0)

print(performance_summary)
```

```python
# Highest Goal Difference (ratio of expected goals to goals scored)
performance_summary['Goal_Difference']=performance_summary['Total_Goals']/performance_summary['Total_xG']
print(performance_summary[['Home','Goal_Difference']].sort_values(by='Goal_Difference', ascending=False))
```

```python
print("""In football, expected goals provides an insight into a team's attacking performance by 
estimating how many goals they should score based on the quality of their created chances.
A team's ratio being close to 1, indicates that they are scoring at an optimal rate.
While overperfomance ( > 1) can be a good indicator, it may also suggest a regression towards the mean as the season goes on.
In our dataset, Luton Town have an xG/G ratio of 1.23, an overperformance that ultimately led to their relegation.
Earlier on, we could also see that Luton were ranked 21st in total xG generated.
Luton Town are an example of overperformance not always being sustainble for a club's success.
In contrast, teams like Manchester City and Arsenal have remained competitive despite showcasing a slight overperformance in their xG, 
highlighting that overperformance in expected goals is not applicable to all teams in the league.""")
```
